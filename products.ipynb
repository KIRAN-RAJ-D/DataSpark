```python
df['Unit Cost USD'] = df['Unit Cost USD'].str.replace('$', '')
df['Unit Price USD'] = df['Unit Price USD'].str.replace('$', '')
df.head()
```

```python
df['Unit Cost USD'] = df['Unit Cost USD'].str.replace(',', '').str.strip().astype(float)
df['Unit Price USD'] = df['Unit Price USD'].str.replace(',', '').str.strip().astype(float)
```

```python
df = df.rename(columns={"ProductKey":"Product_Key","Product Name":"Product_Name","Unit Cost USD":"Unit_Cost","Unit Price USD":"Unit_Price","SubcategoryKey":"Subcategory_Key","CategoryKey":"Category_Key"})
df.head()
```
